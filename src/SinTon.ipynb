```python
from scipy.io import wavfile
import numpy as np
import matplotlib.pyplot as plt

#DFT 
def dft(x):
    N = len(x) # Amount of Signals measured
    n = np.arange(N) # Vector of increasing natural Numbers from 0 to N
    k = n.reshape((N, 1)) # Transpose n; k is N x 1 vector
    M = np.pi * k * n / N # Symmetric N x N - Matrix 
    W = np.exp(-2j * M) 
    
    X = np.dot(W, x) # Matrix multiplication. x is set to a N x 1 - Matrix
    
    return X
```

# Option 2: Create own Sound

```python
fs = 1000 # Abtastrate in Hz
t = np.arange(0, 3, 1/fs)  # Zeitvektor von 0 bis 3 Sekunde
f1 = 1  # Frequenz des Sinussignals in Hz
x = np.sin(2 * np.pi * f1 * t) # Sinussignal

plt.scatter(np.arange(len(x)), x) #plot Signals
```

```python
# Durchfhrung der DFT
X = dft(x)  # DFT des Signals

# Berechnung des Amplituden-Spektrums
amplitude_spectrum = np.abs(X)  # Betrag der komplexen Zahlen

# Erstellung der Frequenzachse ohne ffttreq
freq_own = np.arange(0, 0.5 * fs, fs/len(x))

# Plot des Amplituden-Spektrums
plt.figure(figsize=(10, 5))
plt.plot(freq_own, amplitude_spectrum[:len(freq_own)])
plt.title('Amplitude-Spectrum')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
